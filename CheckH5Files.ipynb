```python
import h5py
import json
import sys
sys.path.append('/weka/prior/mattw/robo_mm_olmo')

from data.utils.vida_utils import parse_local_scene_info, convert_byte_to_string

# Open the first training file
file_path = "/weka/prior/datasets/vida_procthor_with_holodeck_assets/2025_07_15/opentype_tasks/ObjectNavType/train/000000/hdf5_sensors.hdf5"

print(f"Opening: {file_path}\n")

with h5py.File(file_path, 'r') as f:
    # List all episodes
    episodes = list(f.keys())
    print(f"Total episodes: {len(episodes)}")
    print(f"Episode keys: {episodes[:10]}\n")
    
    # Look at first episode
    episode = f[episodes[0]]
    print(f"=== Episode {episodes[0]} ===")
    print(f"\nAll fields in episode:")
    for field in sorted(episode.keys()):
        if hasattr(episode[field], 'shape'):
            print(f"  {field:30s} | shape: {str(episode[field].shape):20s} | dtype: {episode[field].dtype}")
    
    # Show some actual data
    print(f"\n=== Sample Data from Frame 0 ===")
    
    # Scene info
    scene_raw = episode['local_scene_info'][0]
    scene_info = json.loads(parse_local_scene_info(scene_raw))
    print(f"\nScene Info:")
    print(f"  Room type: {scene_info['scene'].get('room_type', 'N/A')}")
    print(f"  Floor color: {scene_info['scene']['floor'].get('color_name', 'N/A')}")
    print(f"  Wall color: {scene_info['scene']['wall'].get('color_name', 'N/A')}")
    
    # Objects
    objects = scene_info.get('objects', [])
    print(f"\n  Objects in frame: {len(objects)}")
    for i, obj in enumerate(objects[:5]):
        print(f"    {i+1}. {obj.get('color_name', 'N/A')} {obj.get('synset', 'N/A')}")
        print(f"       asset_id: {obj.get('asset_id', 'N/A')[:30]}...")
        print(f"       distance: {obj.get('distance', 'N/A')}")
    
    if len(objects) > 5:
        print(f"    ... and {len(objects) - 5} more objects")
    
    # Task info
    try:
        task_spec = convert_byte_to_string(episode['templated_task_spec'][-1], None)
        print(f"\nTask Specification: {task_spec[:200]}...")
    except:
        print("\n(Could not read task spec)")
```
